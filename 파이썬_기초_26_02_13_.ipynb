````python
class Robot:
    def __init__(self): #__init은 함수 호출하면 바로 나오는 애
        self.sentence = "안뇽"
        #self는 만들어진 인스턴스 자기자신 자체를 가리킴.
        #self.무엇은 self 내부의 것(변수, 함수)를 가리킴.
        #클래스 내부의 변수는 attribute라고 한다.

    def greeting(self):
        #print(sentence)라고 하면 error가 뜬다.
        print(self.sentence)


bot1=Robot() #호출
bot1.greeting()
````

````python
class Robot:
    def __init__(self, height, sentence): #밖에서 넘어오는 매개변수는 무조건 self 뒤에
        self.height =height #self.height는 밖에서 들어온 것이다. height랑 다f른 것임. self.height라는건 자기 자신에게 부여하겠다는 의미.
        self.sentence = sentence

    def greeting(self):
        print(self.sentence)
        print(f"제 키는 {self.height}이에요")


bot1 = Robot(180,"안녕") #self에는 값이 들어가지 않는다.
bot1.greeting()

print(bot1.height) #밖에서 속성 확인
print(bot1.sentence) #속성 확인

bot1.height-=10
bot1.greeting()
````

````python
#class 수정 금지로 바꾸는 법

class Robot:
    def __init__(self, height, sentence):
        self.__height = height #비공개로 만들고 싶은 속성은 앞에 밑줄 두개!!!
        self.sentence = sentence

    def greeting(self):

        print(self.sentence)
        self.__height-=10  #class내부에선 비공개 해도 변환시킬 수 있따.
        print(f"제 키는 {self.__height}이에요")


bot1=Robot(180,"hi")
bot1.greeting()
#@print(bot1.height) #보이지 않는다. 밖에서 존재 자체를 알수 없음.
print(bot1.sentence)
# bot1.__height-=10 __height는 비공개라서 바꿀수도 없음.
````

````python
#상속-->효율화


class Robot: #부모 클래스
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot): #자식 클래스, Robot클래스를 상속 받음 --> robot이 # 가지는 함수 사용가능!!
    def study(self):
        print("공부합니다")

bot1=Study_bot()
bot1.greeting() #ㄴstudy_bot 클래스에서는 greeting이 없지만 물려받았기 때문에 사용가능
bot1.study()
````

````python
class Robot: #부모 클래스
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot): #자식 클래스, Robot클래스를 상속 받음 --> robot이 # 가지는 함수 사용가능!!
    def greeting(self):
        print("method overriding success.")

bot1=Study_bot()
bot1.greeting()
````

````python
class Robot: #부모 클래스
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot): #자식 클래스, Robot클래스를 상속 받음 --> robot이 # 가지는 함수 사용가능!!
    def greeting(self):
        super().greeting() #super()가 부모 클래스를 가리킴. #부모의 것을 사용하는 것이 가능하다.
        print("method overriding success.") #부모의 원래 기능은 유지하고 거기에 덧 붙이고 싶을때, 오버라이딩 사용가능.

bot1=Study_bot()
bot1.greeting()
````

````python
class Robot:
    def func1(self):
        print("부모 클래스의 func1이 실행됐다.")
        self.money=100

class Study_bot(Robot):
    def func2(self):
        print("자식의 클래스의 fubc2 실행")
        self.book =10

bot1=Study_bot()
bot1.func1() #이걸 꼭 써야 money에 100 저장 자녀 class에서 부모의 함수를 실행해야 상속이 됨.
bot1.func2()
print(bot1.money) #부모 클래스에 attribute가 저장된다.
print(bot1.book)
#
````

````python
#딥러닝 코드에선 아래와 같은 코드를 많이 보게 된다.


class Robot:
    def __init__(self):
        print("부모 클래스의 init 실행")
        self.money =100

    def test(self):
        print("hi")

class Study_bot(Robot):
    def __init__(self):
        super().__init__()  #이렇게 내부에서 실행해줘야 부모 클래스 attribute를 상속 받는다.
        print("자식 클래스의 init실행")
        self.book =10

bot_parant=Robot()
print(bot_parant.money)

bot_child=Study_bot()
print(bot_child.book)

print(bot_child.money)
bot_child.test()

````

````python
class Robot:
    def greeting(self):
        print("로봇입니다.")


class Life:
    def eat(self):
        print("냡냡")


class Study_bot(Robot, Life):
    def study(self):
        print("열심열심")


bot1=Study_bot() #Robot 클래스(개념)에 해당하는 instance를 만듦.
bot1.greeting()
bot1.eat()
bot1.study()
````
